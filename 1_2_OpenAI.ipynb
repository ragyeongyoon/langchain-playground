```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_openai import OpenAI

llm = OpenAI(model='gpt-4.1-nano')
llm.invoke('최근 회계업계 이슈에 대해서 알려줘')
```

```python
for token in llm.stream('최근 회계업계 이슈에 대해서 알려줘'):
    print(token, end='')
```

3. ChatModel 확인

```python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(model='gpt-4.1-nano')
resp = llm.invoke('최근 회계업계 이슈에 대해서 알려줘')
```

```python
resp.content
```

```python
resp.response_metadata
```

```python
resp = llm.bind(logprobs=True).invoke('최근 회계업계 이슈에 대해서 알려줘')
resp.response_metadata
```

```python
resp = llm.stream('최근 회계업계 이슈에 대해서 알려줘')
for token in resp:
    print(token.text(), end='')
```
